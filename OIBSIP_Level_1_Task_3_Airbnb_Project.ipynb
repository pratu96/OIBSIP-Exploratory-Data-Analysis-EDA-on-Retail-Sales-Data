**Boxplots to detect outliers**

```python
#list of numerical colums
numerical_columns = ['price','minimum_nights','number_of_reviews','reviews_per_month','availability_365']

#plot boxplots
plt.figure(figsize=(15,8))
for i, col in enumerate(numerical_columns, 1):
  plt.subplot(2, 3, i) # Corrected sunplot to subplot
  sns.boxplot(y=df[col])
  plt.title(col)
plt.tight_layout()
plt.show()
```

**Z Score Mean**

```python
from scipy.stats import zscore

#Compute Z-scores
z_scores = np.abs(zscore(df[numerical_columns]))

#set a threshold (e.g., 3 standard deviations)
threshold = 3
outliers_z = (z_scores > threshold).any(axis=1)

#filter the outliers
df_outliers_z = df[outliers_z]
print('Number of outliers detected (z-score):', df_outliers_z.shape[0])
```

**Isolation Forest(Machine Learning-Based)**

```python
from sklearn.ensemble import IsolationForest

#fit isolation forest
iso_forest = IsolationForest(contamination=0.02, random_state=42)
df['outliers_iforest'] = iso_forest.fit_predict(df[numerical_columns])

#count detected outliers
print('Number of outliers detected (isolation forest):', sum(df['outliers_iforest']== -1))
```
